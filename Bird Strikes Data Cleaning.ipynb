```python
# Check for missing values and handle them
df.dropna(inplace=True)  # Drop rows with missing values
```

```python
# Remove any duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Function to extract numeric values from a string
def extract_numeric_value(text):
    numeric_value = re.findall(r'\d+.*\d*', text)
    if numeric_value:
        return numeric_value[0]
    else:
        return None


# Convert columns with numeric values to the appropriate data type
df['Altitude bin'] = df['Altitude bin'].apply(lambda x: extract_numeric_value(str(x)))
df['Wildlife: Number struck'] = df['Wildlife: Number struck'].apply(lambda x: extract_numeric_value(str(x)))
df['Wildlife: Number Struck Actual'] = df['Wildlife: Number Struck Actual'].apply(lambda x: extract_numeric_value(str(x)))
df['Aircraft: Number of engines?'] = df['Aircraft: Number of engines?'].apply(lambda x: extract_numeric_value(str(x)))
df['Cost: Total $'] = df['Cost: Total $'].apply(lambda x: extract_numeric_value(str(x)))
df['Feet above ground'] = df['Feet above ground'].apply(lambda x: extract_numeric_value(str(x)))
df['Number of people injured'] = df['Number of people injured'].apply(lambda x: extract_numeric_value(str(x)))


# Check if the column exists before applying transformation
#if 'Aircraft: Number of engines' in df.columns:
#    df['Aircraft: Number of engines ?'] = df['Aircraft: Number of engines ?'].apply(lambda x: extract_numeric_value(str(x)))
```
